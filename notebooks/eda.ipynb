### **EDA**

```python
# Explore the dataset
def explore_data(df):
    num_rows = df.shape[0]
    num_features = df.shape[1]
    data_types = df.dtypes

    print(f'Number of rows: {num_rows}')
    print(f'Number of features: {num_features}')
    print('\nData Types:')
    print(data_types)

explore_data(df)
```

```python
# EDA
def perform_eda(df):
    duplicates = df.duplicated().sum()
    missing_values = df.isna().sum()

    print(f'Number of duplicate rows: {duplicates}')
    print(f'\nNumber of missing values: {missing_values}')

perform_eda(df)
```

### **Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Format figure in dark mode
plt.style.use('dark_background')

# Create correlation heatmap
def plot_correlation(df):
    fig, ax = plt.subplots(figsize=(15, 6))

    # Define mask for for an upper triangle
    mask = np.triu(np.ones_like(df.corr(), dtype=bool))

    # Plot the heatmap
    sns.heatmap(
        df.corr().round(2),
        ax=ax,
        mask=mask,
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    ax.set_title('Triangle Correlation Heatmap')
    plt.show()

plot_correlation(df)
```

```python
# Ranked correlation heatmap
def ranked_heatmap(df, target_column):
    plt.figure(figsize=(4,8))

    # Plot the heatmap
    heatmap = sns.heatmap(
        df.corr()[[target_column]].sort_values(by=target_column, ascending=False),
        vmin=-1,
        vmax=1,
        annot=True,
        cmap='BrBG'
    )

    heatmap.set_title(f'Features Correlating with {target_column}')
    plt.show()

ranked_heatmap(df, target_column='satisfied')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create histogram and boxplot
def visualize_data(df, column_name):
    fig, axes = plt.subplots(
        figsize=(10, 6),
        nrows=2,
        sharex=False
    )

    sns.set_style('dark')

    # Histogram
    sns.histplot(
        df[column_name],
        ax=axes[0],
        bins='auto',
        kde=True
    )

    axes[0].tick_params(axis='x', rotation=45)

    # Boxplot
    sns.boxplot(
        data=df,
        x=column_name,
        ax=axes[1]
    )

    plt.tight_layout()
    plt.show()

visualize_data(df, column_name='Flight Distance')
```
